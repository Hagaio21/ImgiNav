```python
import torch
from torchvision import models, transforms
from PIL import Image

device = "cuda" if torch.cuda.is_available() else "cpu"
resnet = models.resnet18(weights="IMAGENET1K_V1").to(device)
resnet.fc = torch.nn.Identity()  # remove classifier head

img = Image.open(r"C:\Users\Hagai.LAPTOP-QAG9263N\Desktop\Thesis\repositories\ImagiNav\dataset\00b88e19-d106-4ab8-a322-31c494a0a6b9\rooms\3016\povs\seg\00b88e19-d106-4ab8-a322-31c494a0a6b9_3016_v04_pov_seg.png")
t = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5,0.5,0.5], std=[0.5,0.5,0.5]),
])
x = t(img).unsqueeze(0).to(device)

with torch.no_grad():
    fmap = resnet.avgpool(resnet.layer4(resnet.layer3(resnet.layer2(resnet.layer1(resnet.maxpool(resnet.relu(resnet.bn1(resnet.conv1(x)))))))))
    global_emb = fmap.view(fmap.size(0), -1)  # global vector


print(global_emb)
```
